```python
lp=1
total=40
name=""
sa=0

def userName():
    global name
    name=input("Enter your name :")
    #print("welcome",name)

def printNotes(m):
        a=[500,100,50,20,10,5,2,1]
        temp=0
        i=0; 
        temp=m;
        for i in range(0,8):
            if int(temp/a[i]) != 0 :
                print(a[i]," notes is:",int(temp/a[i]))
            temp=temp%a[i]
    
def takeCash():
    cash=int(input("Enter your Cash :"))
    #print("\b/-")
    global total
    if cash>total :
        printNotes(cash-total)
    
    elif cash<total:
        total-=cash
        print("PLEASE PAY",total,"/- MORE") 
        takeCash()
        
    
def cupsOrPlats(c):
    if c==1 or c==3:
        return "plats:"
    if c==2:
        return "cups:"

def printBill():
    i=1
    print("ok..",name,"Your Bill :",total)
    print("sr.\t|\tItem\t|\tQty\t|\tPrice\t|\tTotal")
    print("------------------------------------------------------------------------------------")
    if sa!=0 :
        print(i,"\t|\tSamosa\t|\t",sa,"\t|\t20\t|\t",sa*20)
        i+=1
        
    print(i,"\t|\tTea\t|\t2\t|\t10\t|\t20")
    i+=1
    print(i,"\t|\tDosa\t|\t2\t|\t10\t|\t20")
    i+=1
    print("------------------------------------------------------------------------------------")
    print("                                                  total :",total)
    takeCash()
    
    
def billing(item,rate):
    global total
    total= total+ item*rate
    
def takingOrder(c,rate):
    pc=cupsOrPlats(c)
    item=int(input("How many "+pc)) 
    billing(item,rate)
    print("ok..Anything else...")
    return item
    
def order(c):
    if c==1:
        print("you select Samosa")
        global sa
        sa+=takingOrder(c,20)
    elif c==2:
        print("you select Tea")
        takingOrder(c,10)
    elif c==3:
        print("you select Dosa")
        takingOrder(c,35)
    elif c==0:
        #print("you select Exit")
        printBill()
        global lp
        lp=0
    else:
        print("Wrong selection")
    
def menu():
    #print("::MENU::\n1.Samosa 20/-\n2.Tea    10/-\n3.Dosa   35/-\n0.Exit\nchoice : ")  
    print('''::MENU::
1.Samosa 20/-
2.Tea    10/-
3.Dosa   35/-
0.Exit
''')
    c= int(input("choice :"))
    #print("your",c)
    order(c)
    

def main():
    print("# Welcome to Asterisc Hotel")
    userName()
    while lp==1:
            menu()
            #print("lp =",lp)
    
```
